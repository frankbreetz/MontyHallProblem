```python
"""
Function
--------
goat_door

Simulate the opening of a "goat door" that doesn't contain the prize,
and is different from the contestants guess

Parameters
----------
prizedoors : array
    The door that the prize is behind in each simulation
guesses : array
    THe door that the contestant guessed in each simulation

Returns
-------
goats : array
    The goat door that is opened for each simulation. Each item is 0, 1, or 2, and is different
    from both prizedoors and guesses

Examples
--------
>>> print goat_door(np.array([0, 1, 2]), np.array([1, 1, 1]))
>>> array([2, 2, 0])
"""
def goat_door(prizedoors,guesses):
    answer=[]
    print(len(prizedoors))
    if(not len(prizedoors)==len(guesses)):
        return "Error: The arrays are of two different sizes"
    else:
        for x in range(0,len(prizedoors)):
            if(prizedoors[x]==guesses[x]):
                if(prizedoors[x]==0):
                    answer.append(1)
                if(prizedoors[x]==1):
                    answer.append(2)
                if(prizedoors[x]==2):
                    answer.append(0)
            else:
                if(prizedoors[x]+guesses[x]==3):
                    answer.append(0)
                elif(prizedoors[x]+guesses[x]==1):
                    answer.append(2)
                elif(prizedoors[x]+guesses[x]==2):
                    answer.append(1)
        return answer
                    
    
```
